```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
#iris
```

```python
dataset = pd.DataFrame(data = iris['data'], columns = iris['feature_names'])
dataset['species'] = iris['target']
#irisのターゲットをdatasetに追加


#データフレームの表示行数指定
pd.set_option('display.max_rows', 150)
#print(dataset)
```

```python
#dataset
```

```python
#dataset.describe()
```

```python
# 目的変数(Y)、説明変数(X)
#目的変数Yが予測対象(species)、説明変数Xが花びらの大きさとなります。
Y = np.array(dataset['species'])
X = np.array(dataset[iris['feature_names']])
```

```python
# データの分割
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

#データを分割する：test_size=0.3 →　全体のデータ数の30%をテストデータとし、70%を訓練用データとする。これをX, Yの両方に適用する。
#random_state: 乱数生成のシードとなる整数または、RandomState インスタンスを設定します。指定しなかった場合は、Numpy のnp.random を用いて乱数をセットします。

# print(X_train)
# print(X_test)
# print(len(X_train)) #当然105
# print(len(X_test)) #当然45
```

```python
model = SVC(gamma='scale')
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(Y_test[:45])
print(Y_pred[:45])
```

```python
metrics.accuracy_score(Y_test, Y_pred)
#正確性評価
```

```python
df=pd.DataFrame(
    {"実際": Y_test[:45], 
     "予測": Y_pred[:45], }
)

#df

#indexの消し方参考記事→https://www.delftstack.com/ja/howto/python-pandas/pandas-remove-index/

    
```

```python
#結果対応のためのカラムの作成

array=[]
for i in range(45):
    if Y_test[i]==Y_pred[i]:
        array.append("")
    else:
        array.append("NG")
        
#print(array)
```

```python
df=pd.DataFrame(
    {"実際": Y_test[:45], 
     "予測": Y_pred[:45], 
     "結果": array}
)

df.style.set_properties(**{'background-color': 'black', # 背景
                           'color': 'lawngreen', # 文字色
                           'border-color': 'white', # 枠の色っぽいが、変わってない？
                           'align':'left'}) # 文字の揃える位置っぽいが、変わってない？
```
