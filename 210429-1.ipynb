```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2)
dx = 0.1  # xの微小な変化
e = np.e  # ネイピア数

y_e = e**x  # 元のべき乗の関数
y_de = (e**(x+dx) - e**x) / dx  # yの微小な変化/xの微小な変化

plt.plot(x, y_e, label="e^x")
plt.plot(x, y_de, label="de")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
class Calc:  # Calcクラス
    def __init__(self, a):  # __init__メソッド
        self.a = a

    def add(self, b):  # addメソッド
        print(self.a + b)

    def multiply(self, b):  # multiplyメソッド
        print(self.a * b)
```

```python
cl = Calc(3)  # インスタンスclを生成
cl.add(4)  # 3 + 4
cl.multiply(4) # 3 × 4
```

```python
class Calc:
    def __init__(self,a):
        self.a=a
    
    def add(self,b):
        print(self.a+b+b)

    def multiply(self,b):
        print(self.a*b*b)
```

```python
cl = Calc(4)
cl.add(5)
cl.multiply(8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Irisデータの読み込み
iris = datasets.load_iris()

# 各花のサイズ
iris_data = iris.data
# print(iris_data)
# print(iris_data.shape)  # 形状

# 散布図で表示
st_data = iris_data[:50]  # Setosa
vc_data = iris_data[50:100]  # Versicolor
plt.scatter(st_data[:, 0], st_data[:, 1], label="Setosa")  # Sepal lengthとSepal width
plt.scatter(vc_data[:, 0], vc_data[:, 1], label="Versicolor")  # Sepal lengthとSepal width
plt.legend()

plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.show()
```
